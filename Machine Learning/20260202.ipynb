```python
import statistics as stats
#import pandas as pd
#import seaborn as sn
li=[22, 20, 21, 22, 24, 28, 26, 30, 27, 28, 5, 31, 29, 70]
q1, q2, q3=stats.quantiles(li)
print(q1, q2, q3)
iqr=q3-q1
low_range=q1-1.5*iqr
high_range=q3+1.5*iqr

for i in li:
    if i<low_range:
        print(i, "lower outlier")
    elif i>high_range:
        print(i, "higher outlier")
```
