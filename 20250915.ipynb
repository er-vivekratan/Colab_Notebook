```python
import numpy as np
import matplotlib.pyplot as plt

am = float(input("Enter am: "))
fc = float(input("Enter fc: "))
fs = float(input("Enter fs: "))

t = np.arange(0, 1, 1/fs)

signal_m = am * np.cos(2 * np.pi * fc * t)

plt.figure(figsize=(10, 4))
plt.plot(t, signal_m)
plt.xlabel('T (s)')
plt.ylabel('Amp')
plt.title('M(t) = am*cos(2*pi*fc*t)')
plt.grid(True)
plt.show()
```
