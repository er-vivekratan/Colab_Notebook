```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
f = 10  # Carrier frequency
fm = 1  # Message frequency
theta = 0

dt = 1e-4
t = np.arange(0, 1, dt)

# Message signal
m_t = np.cos(2 * np.pi * fm * t)

# Carrier signal
c_t = np.cos(2 * np.pi * f * t + theta)

# Double-sided suppressed carrier signal
y = m_t * c_t

plt.plot(t, y)
plt.grid(True)
plt.xlabel('time (sec)')
plt.ylabel('Amplitude (Volts)')
plt.title('Double-sided suppressed carrier signal')
plt.show()

# Fourier Transform
y_fft = np.fft.fft(y)

# Inverse y_fft = np.fft.ift(y1)

freq_axis = np.fft.fftfreq(len(y), d=dt)
plt.plot(freq_axis, np.abs(y_fft) / np.max(np.abs(y_fft))) # Plot magnitude spectrum
plt.xlabel('Freq (Hz)')
plt.ylabel('Normalized Amplitude')
plt.grid(True)
plt.xlim([-20, 20]) # Adjust x-limits to show the frequencies of interest
plt.title('Fourier Transform of Double-sided suppressed carrier signal')
plt.show()
```
