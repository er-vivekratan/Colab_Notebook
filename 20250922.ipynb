```python
import numpy as np
import matplotlib.pyplot as plt


A = 1
theta = 0 #inital phase of car_sig
dt = 1e-4 #sampling interval
t = np.arange(0, 1, dt) #Time Axis
mu = 0.75  # Modulation Index
```

# **Message Signal**

```python
msg_freq = 1
msg_sig = np.cos(2 * np.pi * msg_freq * t)

plt.plot(t, msg_sig)
plt.grid(True)
plt.xlabel('time (s)')
plt.ylabel('Amp (V)')
plt.title('Message Signal')
plt.show()
```

# **Carrier Signal**

```python
car_freq = 100
car_sig = np.cos(2 * np.pi * car_freq * t + theta)

plt.plot(t, car_sig)
plt.grid(True)
plt.xlabel('time (s)')
plt.ylabel('Amp (V)')
plt.title('Carrier Signal')
plt.show()
```

# **Double Sidded Suppressed Carrier Signal**

```python
dsc_sig = msg_sig * car_sig

plt.plot(t, dsc_sig)
plt.grid(True)
plt.xlabel('time (s)')
plt.ylabel('Amp (V)')
plt.title('Double-sided suppressed carrier signal')
plt.show()
```
